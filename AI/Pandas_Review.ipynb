# TUTORIAL 3/11 - Indexes - How to Set, Reset, and Use Indexes ([link of the video](https://www.youtube.com/watch?v=W9XjRYFkkyw&list=PL-osiE80TeTsWmV9i9c58mdDCSskIFdDS&index=3))

As we said previously, the indexes are those unique values which appear at the left hand side of a df when printed. Usually these indexes are integers (0,1,2,3...). However this indexes can also be "other things". In this tutorial we will se how to work with indexes and take benefits of so.

```python
print(people_df['email'])
print('------------------------------------------')
print(people_df.set_index('email')) # This is a way to set a specific column of a df as the indexes of it.
                                    # So in this case, the "email" column is both a column of the people_df but also
                                    # is the indexes of it.
print('------------------------------------------')

# In order to understand better what this is about, let's see the following:

print(people_df.set_index('email').loc['CoreyMSchafer@gmail.com']) # Note that in the first argument of "loc" we now pass an email (instead
                                                                   # an integers). This is because we have set the indexes as the emails.

                                                                   # That means that if we pass an integer in the first argument we will get an error.
                                                                   # However, one can still getting access to the rows of a df by integers using the
                                                                   # "iloc" method.
print('------------------------------------------')
print(people_df.set_index('email').loc['CoreyMSchafer@gmail.com','first'])
```

```python
# One important remark.

print(people_df) # Note that if this df is printed again, it will remain unchangeable (that is, the indexes still being: 0,1,2,3...)
print('------------------------------------------')
# So if when using methods which are intended to change the structure of the df (for instance, the "set_index()" method), it is neccesary
# to pass the argument "inplace = True", in order to to produce a change in the original df.

people_df.set_index('email',inplace = True)
print(people_df) # Now, we can see that once the argument "inplace = True" is given, we will actually change our original df.
                 # When the argument "inplace" isn't passed it is by default "False".
                 # Of course, this argument ("inplace") existis for several other method besides "set_index()".

print('------------------------------------------')

print(people_df.index) # The index attribute returns all the index of the df.

print('------------------------------------------')
people_df.reset_index(inplace = True) # This method reasigns the default indexes of the df (0,1,2,3). How ever, it's important to note that
print(people_df)                      # the order of the columns have changed. (now the "email" column has changed).

# We can also rearrange the columns of our df in the desired order:
print('------------------------------------------')
people_df = people_df[['first', 'last', 'email']] # All we have to do is passing a list with the name of the columns in the desired order.
print(people_df)
```

So far we've seen some methods to change values of multiple rows. However, there are other methods which let us perform more complex operation when we want to change values of multiple rows. These methods are:

- apply.
- applymap
- map
- replace.

We will go over each of them.

Let's see the ***apply*** method.

```python
# This method can be applied over a df or a Series and the behaviour for each o these will be different.

# Let's see what happend when we apply the "apply" method over a Series (df column).

# Basically it applies a function  to each value of the Series and return the new value. The argument of the function
# shall be the old value of the Series. Let's see how it works:

print(people_df['EMAIL'].apply(len)) # With this instruction we passed the built-in function "len" to the "apply" method
                                     # over the series people_df['EMAIL']. So what it returns is another series with the lenght
                                     # of the string (in this cases the emails) of each person.
                                
print('------------------------------------------')

# However, we can also pass a function created by us (this is in fact the powerful feature of the "apply" method). Let's see:

def upper_case_email (email): # Of corse this is a very simple function, but it can be as complicated as we want.
    return email.upper()

print(people_df['EMAIL'].apply(upper_case_email)) # As we can see, when we pass our "upper_case_email" function, to the apply method
                                                  # we get a Series with all the emails but in sharp uppercase (mayúscula sostenida). 

                                                  # Again, this is a very simple function, but it can be as complex as we want.

print('------------------------------------------')

 # We can also pass lamda functions (for instance, when the function isn't too complicated):


print(people_df['EMAIL'].apply(lambda x: x.split('@')[0] + '@stanford.edu.co')) # In this cases we are using a lambda function which the domain of the
                                                                                # emails (@stanford.edu.co)


# THE BIG ADVANTAGE OF USING THE "apply" METHOD IS THAT WE CAN USE IT INTEAD OF A "for loop" WHICH MIGHT BE COMPUTACIONAL EXPENSIVE (SO SLOW).       

# KEEP THAT CAN ALSO PASS MORE ARGUMENT BESIDES THE ELMENT OF THE SERIES TO THE FUNCTION PASSED TO THE "apply" METHOD (see the documentation:
# https://pandas.pydata.org/docs/reference/api/pandas.Series.apply.html).
```

```python
 # In the case that we want to update our "EMAIL" column of our data frame, all that we have to do is:      

people_df['EMAIL'] =   people_df['EMAIL'].apply(lambda x: x.split('@')[0] + '@stanford.edu.co') 
people_df
```

```python
# Now let's look at how "apply" works when we apply it to a df intead of a Series (Column of the df).

# In this case the functions (built-in or ours) passed to the apply method takes as the argument every single column of our df. Let's see
# some examples.

print(people_df.apply(len , axis = "rows")) # As we can see, it returns a  series with the length of each column (or Series) of the df.

print('------------------------------------------')
# However, if pass "axis = 'columns'" then works over each rows of our df. Let's see:
print(people_df.apply(len , axis = "columns"))

print('------------------------------------------')

# Let's see another example:

print(people_df.apply(pd.Series.min)) # In this case it returns a Series with the minimum value of each column
                                      # of our df (in this case, because all the values in our df are string the minimum will
                                      # be taken based on the alphabetical order).

# Let's do the same but now using lamda notation:
print('------------------------------------------')
print(people_df.apply(lambda x: x.min()))

```

Now, let's take a look to the ***applymap*** method:

This method only works over a df (not Series as "apply" does).

When we use the ***applymap*** over a df, the function passed to the "applymap" will receive each value of our df.

The ***applymap*** only returns a df.

Let's see an example:

```python
people_df.applymap(len) # This returns a df with the length of each value of our df.
```

```python
people_df.applymap(str.lower)
```

Now, let's take a look to the ***map*** method.

This method only works over a Series and it's meant to substitute (or map) a value on a Series with another value. Let's see some examples:

```python
people_df['First_Name'].map({'Corey': 'Chris', 'Jane': 'mary'}) # Note that the keys that we didn't pass to the dictinary
                                                                # are substituted (or mapped) to a "NaN" value.
```

Finally, we have the ***replace*** method, which works pretty much the same that the ***map*** method, expect that keys which aren't provided in the dictionary are mapped to the same value in the returned Series:

```python
people_df['First_Name'].replace({'Corey': 'Chris', 'Jane': 'Mary'}) # This time, the "Jhon" value was not mapped to a "NaN" value,
                                                                    # instead it remains the same in the returned Series.
```
